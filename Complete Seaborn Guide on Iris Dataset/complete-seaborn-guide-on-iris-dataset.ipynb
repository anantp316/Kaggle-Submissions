# What is Seaborn ?

Seaborn is a Python data visualization library based on matplotlib. Seaborn is a great choice for exploratory data analysis. 

# Why use Seaborn ? Doesn't Matplotlib suffice ?

As mentioned on Seaborn’s official website,

If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.



# Demonstration

For our entire demonstration we will be using the Iris dataset.

**Loading the dataset**

```python
from sklearn.datasets import load_iris
```

```python
dataframe = load_iris()
```

```python
dataframe
```

**Converting the dataset into usable format**

```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame(data= np.c_[dataframe['data'], dataframe['target']],
                  columns= dataframe['feature_names'] + ['Species'])
```

```python
dataframe
```

# Importing Required Libraries

We will import Seaborn as well as Matplotlib for additional customization.

We make use of the Seaborn libraries figure aesthtics to set the theme to a dark-grid.

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("dark")
```

# 1. Scatterplot

A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are coded, one additional variable can be displayed.

We will begin a simple scatterplot of Sepal length against Sepal Width.


```python
sns.lmplot(x='sepal length (cm)', y='sepal width (cm)', data=dataframe)
```

**Adding more aesthetics using the Seaborn parameters**

We get rid of the regresion line that was there by default.

We set the hue of the points by the species. 

```python
sns.lmplot(x='sepal length (cm)', y='sepal width (cm)', data=dataframe,
           fit_reg=False, # No regression line
           hue='Species')   # Color by species
```

# 2. Boxplot 

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are.

We begin by making a simple boxplot

```python
plt.figure(figsize=(10,8)) 
sns.boxplot(data=dataframe)
```

**Getting rid of target variable that is Species**

```python
dataframe_1 = dataframe.drop(['Species'], axis=1)

plt.figure(figsize=(10,8)) 
sns.boxplot(data=dataframe_1)
```

# Understanding the Seaborn Color Palette

Seaborn allows us to set custom color palettes. 

We can simply create an ordered Python list of color hex values.

```python
species_colors = ['#78C850',  # Setosa
                    '#F08030',  # Versicolor
                    '#6890F0',  # Verginica
                  ]
```

# 3. Violin Plot

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. 

Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable. 

Here we plot the Violin Plot of the entire dataset whilst making use of the defined palette.

```python
plt.figure(figsize=(10,8)) 
sns.violinplot( data=dataframe, 
               palette=species_colors) # Set color palette
```

# 4. Swarm Plot

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes. 

Here we create a Swarm plot of Petal length across the species.

```python
plt.figure(figsize=(10,8)) 
sns.swarmplot(y='petal length (cm)', x='Species', data=dataframe, 
              palette=species_colors)
```

# Combining plots

Let us combine our Violin plot and Swarm plot.

```python
plt.figure(figsize=(10,8))
 
sns.violinplot(x='Species',
               y='petal length (cm)', 
               data=dataframe, 
               inner=None, # Remove the bars inside the violins
               palette=species_colors)
 
sns.swarmplot(y='petal length (cm)', 
              x='Species', 
              data=dataframe, 
              color='k', # Make points black
              alpha=0.7) # and slight transparency
 
# Set title with matplotlib
plt.title('Combined Plot')
```

# 5. Correlation Heatmap

**What is correlation?**

Correlation is a statistical measure (expressed as a number) that describes the size and direction of a relationship between two or more variables.

In simple words, it means directionality of 2 variables.

First we calculate the correlation then we plot it.

```python
corr = dataframe_1.corr()
```

**The Plot**

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr)
```

# 6. Histogram

A histogram is a bar graph-like representation of data that buckets a range of classes into columns along the horizontal x-axis.


```python
plt.figure(figsize=(10,8))
sns.histplot(dataframe, x='petal length (cm)',
             kde=True)
```

# 7. Bar Plot

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart. 

**Count Plot is Bar Plot**

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Species', data=dataframe, palette=species_colors)
plt.xticks(rotation=-45) #Rotating X axis labels
```

# 8. CatPlot

Catplot is a relatively new addition to Seaborn that simplifies plotting that involves categorical variables.

```python
g = sns.catplot(x='Species', 
                   y='petal width (cm)', 
                   data=dataframe, 
                   hue='Species',  # Color by species
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

# 9. Density Plot

A density plot is a representation of the distribution of a numeric variable that uses a kernel density estimate to show the probability density function of the variable.



```python
sns.kdeplot(x='sepal length (cm)', y='sepal width (cm)', data=dataframe)
```

# 10. Joint Distribution Plot

Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distributions.

```python
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=dataframe)
```

**Hence we saw how Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.**
